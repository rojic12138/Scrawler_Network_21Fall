示例

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.wait import WebDriverWait

browser=webdriver.Chrome()
try:
    browser.get('https://www.google.com.hk/')
    input=browser.find_element_by_name('q')
    input.send_keys('好耶')
    input.send_keys(Keys.ENTER)
    wait=WebDriverWait(browser,10)
    print(browser.current_url)
    print(browser.page_source)
finally:
    browser.close()
    
```

可以看到自动打开了一个浏览器，进行输入、按enter的步骤，并输出源代码

### Selenium的使用

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
#声明浏览器对象，我只用了谷歌
browser=webdriver.Chrome()
#访问页面
browser.get('https://www.taobao.com/')
#print(brower.page_source)

#查找节点
input_=browser.find_element_by_id('q')
    #还有find_element_by_id name xpath tag_name 等
    #多个节点，将element改成elements
#节点交互
#输入文字send_keys() 清空文字clear() 点击按钮click()
input_.send_keys('海子姐卫衣')
print(input_)

#input_.clear()
brower.close()
```

```python
#动作链
#输入、点击都是针对特定对象的
#有些操作没有特定对象，比如鼠标拖拽，需要用动作链来实现
from selenium.webdriver import ActionChains

browser=webdriver.Chrome()
url='https://www.runoob.com/try/try.php?filename=jqueryui-api-droppable'
browser.get(url)
browser.switch_to.frame('iframeResult')#在源代码里搜frame找到名字
source=browser.find_element_by_id('draggable')
target=browser.find_element_by_id('droppable')
actions=ActionChains(browser)
actions.drag_and_drop(source,target)#瞬移，没有中间路径
actions.perform()
```

```python
#获取节点信息
#通过find_element得到的节点都是WebElement类型，提供了许多接口

src=browser.find_element_by_name('s')
print(src)
print(src.get_attribute('class'))
print(src.text)
print(src.id)
print(src.location)
print(src.tag_name)
print(src.size)
```

```python
#等待：设置一个最长等待时间，如果在规定时间内加载出了这个节点就返回这个节点，否则抛出超时异常
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
browser=webdriver.Chrome()
browser.get('https://www.bilibili.com/')
wait=WebDriverWait(browser,10)
search=wait.until(EC.presence_of_element_located((By.ID,'nav_searchform')))
print(search)
#等待条件还有很多，用到再说
```

```python
#前进和后退
import time
browser.get('https://live.bilibili.com/')
time.sleep(5)
browser.back()#后退
time.sleep(5)
browser.forward()#前进
```

```python
#Cookies
browser=webdriver.Chrome()
browser.get('https://live.bilibili.com/')
print(browser.get_cookies())#获取cookies
browser.add_cookie({'name':'rojic','domain':'.bilibili.com','value':'rojic'})#添加cookies
print(browser.get_cookies())#获取cookies
browser.delete_all_cookies()#删除cookies
print(browser.get_cookies())
```
